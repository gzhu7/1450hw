```python
eruptions = pd.read_csv('C://Users//47129//Desktop//geyser_csv//eruptions.csv',header=None)
waiting = pd.read_csv('C://Users//47129//Desktop//geyser_csv//waiting.csv',header=None)
eruptions.columns=['S']
eruptions
eruptions['X_long'] = eruptions.S.apply(lambda s: 1 if s > 3.25 else 0)
eruptions
```

```python
waiting.columns=['S']
waiting
waiting['Y_long'] = waiting.S.apply(lambda t: 1 if t > 68 else 0)
waiting
```

```python
com = pd.concat([eruptions, waiting], axis=1)
com
```

```python
def X1Y1(a, b):
    if a==1 and b==1:
        return 1
    else:
        return 0
com['X1Y1'] = com.apply(lambda x: X1Y1(x.X_long, x.Y_long), axis = 1)
com
```

```python
def X1Y0(a, b):
    if a==1 and b==0:
        return 1
    else:
        return 0
com['X1Y0'] = com.apply(lambda x: X1Y0(x.X_long, x.Y_long), axis = 1)
com
```

```python
def X0Y0(a, b):
    if a==0 and b==0:
        return 1
    else:
        return 0
com['X0Y0'] = com.apply(lambda x: X0Y0(x.X_long, x.Y_long), axis = 1)
com
```

```python
def X0Y1(a, b):
    if a==0 and b==1:
        return 1
    else:
        return 0
com['X0Y1'] = com.apply(lambda x: X0Y1(x.X_long, x.Y_long), axis = 1)
com
```

```python
X1Y1=com['X1Y1'].sum()
X1Y0=com['X1Y0'].sum()
X0Y0=com['X0Y0'].sum()
X0Y1=com['X0Y1'].sum()
```
